```python
# For Loop
n = 4
for i in range(0, n):
    print(i)
```

```python
# Iterating over a list, tuple, string, dict, set
li = ["rahul", "for", "chauhan"]
for x in li:
    print(x)
    
tup = ("rahul", "for", "chauhan")
for x in tup:
    print(x)
    
s = "abc"
for x in s:
    print(x)
    
d = dict({'x':123, 'y':354})
for x in d:
    print("%s  %d" % (x, d[x]))
    
set1 = {10, 30, 20}
for x in set1:
    print(x),
```

```python
# Iterating by index of sequence 
li = ["rahul", "for", "chauhan"]
for index in range(len(li)):
    print(li[index])
```

```python
# While Loop
cnt = 0
while (cnt < 3):
    cnt = cnt + 1
    print("Hello Rahul")
```

```python
# Continue Control
for letter in 'rahulchauhan':
    if letter == 'a' or letter == 'h':
        continue
    print('Current Letter :', letter)
```

```python
# Break Control
for letter in 'rahulchauhan':
    if letter == 'l' or letter == 'n':
        break

print('Current Letter :', letter)
```

```python
# Pass control to create empty loops
for letter in 'rahulchauhan':
    pass
print('Last Letter :', letter)
```
