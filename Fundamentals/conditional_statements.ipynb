```python
# If Statement
age = 20
if age >= 18:
    print("Eligible to vote.")
```

```python
# If Else Statement
age = 10
if age <= 12:
    print("Travel for free.")
else:
    print("Pay for ticket.")
```

```python
# elif Statement
age = 25

if age <= 12:
    print("Child.")
elif age <= 19:
    print("Teenager.")
elif age <= 35:
    print("Young adult.")
else:
    print("Adult.")
```

```python
# Nested If Else
age = 70
is_member = True

if age >= 60:
    if is_member:
        print("30% senior discount!")
    else:
        print("20% senior discount.")
else:
    print("Not eligible for a senior discount.")
```

```python
# Ternary Condition Statement
age = 20
s = "Adult" if age >= 18 else "Minor"
print(s)
```

```python
# Match Case 
number = 2

match number:
    case 1:
        print("One")
    case 2 | 3:
        print("Two or Three")
    case _:
        print("Other number")
```
