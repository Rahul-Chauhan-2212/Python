# Type of function argument

```python
# Default Arguments
def myFun(x, y=50):
    print("x: ", x)
    print("y: ", y)

myFun(10)
```

```python
# Keyword Arguments
def student(fname, lname):
    print(fname, lname)

student(fname='Rahul', lname='Chauhan')
student(lname='K', fname='Chauhan') 
# Values are passed explicitely so order does not matter
```

```python
# Positional Arguments
def nameAge(name, age):
    print("Hi, I am", name)
    print("My age is ", age)

print("Case-1:")
nameAge("Rahul", 27)

print("\nCase-2:")
nameAge(27, "Rahul")
# Here these need to be passed in correct order
```

```python
# Arbitrary Arguments
def myFun(*args, **kwargs):
    print("Non-Keyword Arguments (*args):")
    for arg in args:
        print(arg)

    print("\nKeyword Arguments (**kwargs):")
    for key, value in kwargs.items():
        print(f"{key} == {value}")

# Function call with both types of arguments
myFun('Hey', 'Welcome', first='Rahul', mid='for', last='Chauhan')
```

```python
# Function within function
def f1():
    s = 'I am rahul'
    def f2():
        print(s)
        
    f2()
f1()
```

```python
# Anonymous Function
def cube(x): return x*x*x   # without lambda
cube_l = lambda x : x*x*x  # with lambda

print(cube(7))
print(cube_l(7))
```

```python
# Return Statement in Function
def square_value(num):
    """This function returns the square
    value of the entered number"""
    return num**2

print(square_value(2))
print(square_value(-4))
```

```python
# Pass by value and Pass by reference
# Function modifies the first element of list
def myFun(x):
    x[0] = 20

lst = [10, 11, 12, 13]
myFun(lst)
print(lst)   # list is modified

# Function tries to modify an integer
def myFun2(x):
    x = 20

a = 10
myFun2(a)
print(a)     # integer is not modified
```

```python
# Recursive Function
def factorial(n):
    if n == 0:  
        return 1
    else:
        return n * factorial(n - 1) 
      
print(factorial(4))
```
