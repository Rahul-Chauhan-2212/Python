```python
# Creating a String
s1 = 'Rahul'  # single quote
s2 = "Rahul"  # double quote
print(s1)
print(s2)
```

```python
# Multi Line Strings
s = """I am Learning
Python """
print(s)

s = '''I'm a 
Rahul Chauhan'''
print(s)
```

```python
# Accessing characters in Strings
s = "Rahul Chauhan"
print(s[0])   # first character
print(s[4])   # 5th character
```

```python
s = "Rahul Chauhan"
print(s[-10])   # 4th character
print(s[-5])    # 5th character from end
```

```python
# Slicing
s = "Rahul Chauhan"
print(s[1:4])    # characters from index 1 to 3
print(s[:3])     # from start to index 2
print(s[3:])     # from index 3 to end
print(s[::-1])   # reverse string
```

```python
# String Iteration
s = "Python"
for char in s:
    print(char)
```

```python
# String Immutability
s = "Rahul Chauhan"
s = "I am " + s[1:]   # create new string
print(s)
```

```python
# Delete a String
s = "Rahul"
del s
```

```python
# Updating a String
s = "rahul Chauhan"
s1 = "R" + s[1:]                   # update first character
s2 = s.replace("rahul", "Rahul")  # replace word
print(s1)
print(s2)
```

```python
# Length function
s = "Rahul Chauhan"
print(len(s))
```

```python
# Upper case and Lower case
s = "Hello World"
print(s.upper())
print(s.lower())
```

```python
# Strip and Replace
s = "   Rahul   " # removes trailing and leading whitespaces
print(s.strip())    

s = "Python is fun"
print(s.replace("fun", "awesome")) # Replace all occurence of this string with new string
```

```python
# Concatenating
s1 = "Hello"
s2 = "World"
print(s1 + " " + s2)
```

```python
# Repeat a string multiple times
s = "Hello "
print(s * 3)
```

```python
# Formatting of String
# Using f-Strings
name = "Rahul"
age = 22
print(f"Name: {name}, Age: {age}")
```

```python
# Using format method
s = "My name is {} and I am {} years old.".format("Rahul", 22)
print(s)
```

```python
# String Memberships
s = "Rahul Chauhan"
print("Rahul" in s)
print("rahul" in s)
```
