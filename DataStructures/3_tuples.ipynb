<h5>A tuple in Python is an immutable ordered collection of elements.</h5>

<ul>
<li>Tuples are similar to lists, but unlike lists, they cannot be changed after their creation (i.e., they are immutable).</li>
<li>Tuples can hold elements of different data types.</li>
<li>The main characteristics of tuples are being ordered, heterogeneous and immutable.</li>
</ul>

```python
# Creating a Tuple
tup = ()
print(tup)

# Using String
tup = ('Rahul', 'Chauhan')
print(tup)

# Using List
li = [1, 2, 4, 5, 6]
print(tuple(li))

# Using Built-in Function
tup = tuple('Python')
print(tup)
```

```python
# Tuple with other data types
tup = (5, 'Welcome', 7, 'Rahul')
print(tup)

# Creating a Tuple with nested tuples
tup1 = (0, 1, 2, 3)
tup2 = ('python', 'code')
tup3 = (tup1, tup2)
print(tup3)

# Creating a Tuple with repetition
tup1 = ('Rahul',) * 3
print(tup1)

# Creating a Tuple with the use of loop
tup = ('Chauhan')
n = 5
for i in range(int(n)):
    tup = (tup,)
    print(tup)
```

```python
# Accessing tuple elements
# Accessing Tuple with Indexing
tup = tuple("Rahul")
print(tup[0])

# Accessing a range of elements using slicing
print(tup[1:4])  
print(tup[:3])

# Tuple unpacking
tup = ("Rahul", "Chauhan", "Python")

# This line unpack values of Tuple1
a, b, c = tup
print(a)
print(b)
print(c)
```

```python
# Concatenation of tuples
tup1 = (0, 1, 2, 3)
tup2 = ('Rahul', 'Chauhan', 'Python')

tup3 = tup1 + tup2
print(tup3)
```

```python
# Slicing Tuples
tup = tuple('RAHULCHAUHAN')

# Removing First element
print(tup[1:])

# Reversing the Tuple
print(tup[::-1])

# Printing elements of a Range
print(tup[4:9])
```

```python
# Delete a Tuple
tup = (0, 1, 2, 3, 4)
del tup

print(tup)
```

```python
# Tuple unpacking with *
tup = (1, 2, 3, 4, 5)

a, *b, c = tup

print(a) 
print(b) 
print(c)
```
